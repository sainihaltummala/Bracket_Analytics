```python
new_2 = pd.DataFrame()
new_2= fileRead.drop('Team',1).dropna()
```

```python
new_2
```

```python
columns = ['gameid','Win?','MP','Index','PTS','Opp PTS']
new_final= new_2.drop(columns, axis=1).dropna()
```

```python
new_final
```

```python
nf = 5
pca = PCA(n_components=nf)
# X is the matrix transposed (n samples on the rows, m features on the columns)
pca.fit(new_final)

X_new = pca.transform(new_final)
```

```python
 print(pca.explained_variance_ratio_) 
var_ration = pca.explained_variance_ratio_
plt.plot(var_ration)
```

```python
pca_components = pd.DataFrame()
pca_components['features'] = new_final.columns
pca_components['pca_contrib'] = pca.components_[0]
pca_components.sort_values('pca_contrib',ascending=False)
```
